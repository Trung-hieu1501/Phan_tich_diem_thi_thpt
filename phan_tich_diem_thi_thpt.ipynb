##Load dữ liệu

```python
# Dữ liệu điểm thi các năm
df_21 = pd.read_csv('/content/drive/MyDrive/data_thpt/diem_thi_2021.csv')
df_22 = pd.read_csv('/content/drive/MyDrive/data_thpt/diem_thi_2022.csv')
df_23 = pd.read_csv('/content/drive/MyDrive/data_thpt/diem_thi_2023.csv')
df_24 = pd.read_csv('/content/drive/MyDrive/data_thpt/diem_thi_2024.csv')
```

##Transform dữ liệu

```python
data = pd.concat([df_21, df_22, df_23, df_24], ignore_index=True)
data
```

```python
data.info()
```

```python
data['code']= data['sbd'].apply(lambda x: int(x/1000000))   # Tạo mã tỉnh
```

```python
# Thêm dữ liệu tỉnh thành vào data
data = pd.merge(data, data_tinh, left_on = 'code', right_on = 'ma_cum', how = 'inner')
data.rename(columns = {'cum':'province'}, inplace = True)
data['province'] = data['province'].str.replace('Sở GDĐT ','')
```

#Phân tích

## Tỷ lệ tăng trưởng thí sinh

```python
student_per_year = data['nam'].value_counts().sort_index()  # Số lượng thí sinh mỗi năm
growth_rate = student_per_year.pct_change()*100   # Tăng trưởng thí sinh từng năm
student_growth_rate = pd.DataFrame({'so_luong_thi_sinh': student_per_year, 'ty_le_tang_truong (%)': growth_rate}).reset_index().rename(columns = {'index': 'nam'})
student_growth_rate.head()
```

```python
fig, ax = plt.subplots(2, 1, figsize=(6, 6))

# Biểu đồ đầu tiên: Số lượng thí sinh
bars1 = ax[0].bar(student_growth_rate['nam'], student_growth_rate['so_luong_thi_sinh'])
ax[0].set_xlabel('Năm')
ax[0].set_ylabel('Số lượng thi sinh')
ax[0].set_title('Số lượng thí sinh các năm')

# Hiển thị giá trị trên từng cột
for bar in bars1:
    yval = bar.get_height()
    ax[0].text(bar.get_x() + bar.get_width()/2, yval, int(yval), ha='center', va='bottom')

# Chỉ hiển thị các giá trị của 'nam' trên trục x
ax[0].set_xticks(student_growth_rate['nam'])
ax[0].set_xticklabels(student_growth_rate['nam'])

# Biểu đồ thứ hai: Tỷ lệ tăng trưởng
#filtered_data = student_growth_rate.dropna(subset=['ty_le_tang_truong (%)'])
bars2 = ax[1].bar(student_growth_rate['nam'], student_growth_rate['ty_le_tang_truong (%)'])
ax[1].set_xlabel('Năm')
ax[1].set_ylabel('Tỷ lệ tăng trưởng (%)')
ax[1].set_title('Tỷ lệ tăng trưởng thí sinh (%)')

# Hiển thị giá trị trên từng cột
for bar in bars2:
    yval = bar.get_height()
    ax[1].text(bar.get_x() + bar.get_width()/2, yval, f"{yval:.2f}%", ha='center', va='bottom')
# Chỉ hiển thị các giá trị của 'nam' trên trục x
ax[1].set_xticks(student_growth_rate['nam'])
ax[1].set_xticklabels(student_growth_rate['nam'])


plt.tight_layout()
plt.show()
```

## Điểm thi trung bình các môn qua các năm

```python
mon_hoc = ['toan', 'van', 'ngoai_ngu', 'vat_ly', 'hoa', 'sinh', 'lich_su','dia_ly', 'gdcd']
```

```python
mean_score_year = data.groupby('nam')[mon_hoc].mean().reset_index().rename(columns = {'index': 'nam'})
mean_score_year
```

```python

plt.figure(figsize=(10, 6))
for mon in mon_hoc:
    plt.plot(mean_score_year['nam'], mean_score_year[mon], marker='o', label=mon)



plt.title('Điểm trung bình các môn Văn, Toán, Anh qua các năm')
plt.xlabel('Năm')
plt.ylabel('Điểm trung bình')
plt.xticks(mean_score_year['nam'])  # Đặt nhãn cho trục x
plt.legend()
plt.grid()

# Hiển thị biểu đồ
plt.tight_layout()
plt.show()
```

- Điểm trung bình các môn văn, sử, địa có xu hướng tăng trong khoảng thời gian từ 2021-2024
- GDCD là môn có điểm trung bình cao nhất trong các năm


##Số lượng thí sinh thi từng môn và điểm thi trung bình của từng năm

```python
student_per_subject = data.groupby('nam')[mon_hoc].count().reset_index().rename(columns = {'index':'nam'})
student_per_subject
```

```python
for i in student_per_subject['nam']:
    plt.figure(figsize=(10, 6))

    # Tạo trục chính
    ax1 = plt.gca()

    # Biểu đồ cột cho số lượng thí sinh
    ax1.bar(mon_hoc, student_per_subject[student_per_subject['nam'] == i].iloc[0, 1:], color='lightblue', label='Số lượng thí sinh')

  # Tạo trục phụ
    ax2 = ax1.twinx()

    # Biểu đồ đường cho điểm trung bình
    ax2.plot(mon_hoc, mean_score_year[mean_score_year['nam'] == i].iloc[0, 1:], marker='o', color='orange', label='Điểm trung bình')

    ax1.set_title(f'Số lượng thí sinh từng môn năm {i}')
    ax1.set_xlabel('Môn học')
    ax1.set_ylabel('Số lượng thí sinh')
    ax2.set_ylabel('Điểm trung bình')

    plt.xticks(mon_hoc)
    ax1.grid()

    # Hiển thị biểu đồ
    plt.tight_layout()
    plt.show()
```

## Phân tích điểm từng khối

```python
data
```

```python
df_filter = data[0:]
df_filter['A'] = df_filter[['toan', 'vat_ly', 'hoa']].sum(axis=1,skipna=False)
df_filter['A1'] = df_filter[['toan', 'vat_ly', 'ngoai_ngu']].sum(axis=1,skipna=False)
df_filter['B'] = df_filter[['toan', 'sinh', 'hoa']].sum(axis=1,skipna=False)
df_filter['C'] = df_filter[['van', 'lich_su', 'dia_ly']].sum(axis=1,skipna=False)
df_filter['D'] = df_filter[['toan', 'van', 'ngoai_ngu']].sum(axis=1,skipna=False)
df_filter.drop(columns = {'sbd', 'toan', 'van', 'ngoai_ngu', 'vat_ly', 'hoa', 'sinh', 'lich_su','dia_ly', 'gdcd', 'code','ma_cum'}, inplace = True)
```

```python
df_filter.head()
```

```python
df_filter_mean_year = df_filter.groupby('nam')[['A','A1', 'B', 'C','D']].mean().reset_index().rename(columns = {'index': 'nam'})

df_filter_count_year = df_filter.groupby('nam')[['A','A1', 'B', 'C','D']].count().reset_index().rename(columns = {'index': 'nam'})

print(df_filter_count_year.head())

print(df_filter_mean_year.head())
```

```python
khoi = ['A', 'A1', 'B','C', 'D']
for i in df_filter_mean_year['nam']:
    plt.figure(figsize=(10, 6))

    # Tạo trục chính
    ax1 = plt.gca()

    # Biểu đồ cột cho số lượng thí sinh
    ax1.bar(khoi, df_filter_count_year[df_filter_count_year['nam'] == i].iloc[0, 1:], color='lightblue', label='Số lượng thí sinh')

  # Tạo trục phụ
    ax2 = ax1.twinx()

    # Biểu đồ đường cho điểm trung bình
    ax2.plot(khoi, df_filter_mean_year[df_filter_mean_year['nam'] == i].iloc[0, 1:], marker='o', color='orange', label='Điểm trung bình')


    ax1.set_title(f'Năm {i}')
    ax1.set_xlabel('Khối thi')
    ax1.set_ylabel('Số lượng thí sinh')
    ax2.set_ylabel('Điểm trung bình')

    plt.xticks(khoi)
    ax1.grid()

    # Hiển thị biểu đồ
    plt.tight_layout()
    plt.show()
```
